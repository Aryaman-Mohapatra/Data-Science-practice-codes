1.Considering weekends and holidays, what was the average daily change in Apple's adjusted closing price throughout the entire data range?

```python
daily_change = apple['Adj Close'].diff()  # Difference between consecutive rows

# Identify weekdays (assuming weekends are Saturday & Sunday)
def is_weekday(date):
  return date.weekday() not in [5, 6]  # 5 - Saturday, 6 - Sunday
weekdays = apple[daily_change.index.map(is_weekday)]  # Filter data for weekdays

# Calculate average daily change on weekdays
average_daily_change = weekdays['Adj Close'].mean()

# Print the answer (considering only weekdays for this example)
print(f"Average daily change in Apple's adjusted closing price on weekdays: {average_daily_change:.2f}")
```
